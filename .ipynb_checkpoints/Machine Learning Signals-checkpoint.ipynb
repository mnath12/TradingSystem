```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks

plt.style.use("seaborn-v0_8")
np.random.seed(42)
tf.random.set_seed(42)
```

```
PATH = "aapl_intraday_data_cleaned.csv"  # change to your file

df = pd.read_csv(PATH, parse_dates=True, index_col=0)
df = df.sort_index()

print(df.shape)
df.head()
```

```
def make_daily_ohlcv_and_rv(df_intraday: pd.DataFrame) -> pd.DataFrame:
    # Daily OHLCV from intraday
    daily = df_intraday.resample("1D").agg({
        "Open": "first",
        "High": "max",
        "Low": "min",
        "Close": "last",
        "Volume": "sum"
    }).dropna()

    # Intraday returns for RV (simple returns)
    intraday_ret = df_intraday["Close"].pct_change().dropna()

    # Daily RV: sqrt(sum r^2 within day)
    rv = intraday_ret.pow(2).groupby(intraday_ret.index.floor("D")).sum().pipe(np.sqrt)
    rv.name = "RV"

    out = daily.join(rv, how="inner").dropna()
    return out

df_daily = make_daily_ohlcv_and_rv(df)
print(df_daily.shape)
df_daily.head()
```

```
df_daily["ret_1"] = df_daily["Close"].pct_change()
df_daily["ret_5"] = df_daily["Close"].pct_change(5)

# Simple momentum and range proxies
df_daily["mom_5"] = df_daily["Close"] / df_daily["Close"].rolling(5).mean() - 1
df_daily["mom_10"] = df_daily["Close"] / df_daily["Close"].rolling(10).mean() - 1
df_daily["range_pct"] = (df_daily["High"] - df_daily["Low"]) / df_daily["Close"]

# Volume surprise
df_daily["vol_ratio"] = df_daily["Volume"] / df_daily["Volume"].rolling(10).mean()

# Lagged RV features (RV is persistent)
df_daily["rv_1"] = df_daily["RV"]
df_daily["rv_5"] = df_daily["RV"].rolling(5).mean()

# Targets (next-day)
df_daily["y_ret"] = df_daily["ret_1"].shift(-1)
df_daily["y_rv"]  = df_daily["RV"].shift(-1)

df_daily = df_daily.dropna()

feature_cols = ["ret_1", "ret_5", "mom_5", "mom_10", "range_pct", "vol_ratio", "rv_1", "rv_5"]
X_df = df_daily[feature_cols].copy()
Y_df = df_daily[["y_ret", "y_rv"]].copy()

print(X_df.shape, Y_df.shape)
X_df.head(), Y_df.head()
```

```
train_frac = 0.7
n = len(df_daily)
cut = int(n * train_frac)

X_train_df = X_df.iloc[:cut]
X_test_df  = X_df.iloc[cut:]

Y_train_df = Y_df.iloc[:cut]
Y_test_df  = Y_df.iloc[cut:]

print("Train:", X_train_df.index.min(), "→", X_train_df.index.max(), len(X_train_df))
print("Test :", X_test_df.index.min(),  "→", X_test_df.index.max(),  len(X_test_df))
```

```
x_scaler = StandardScaler()
yret_scaler = StandardScaler()
yrv_scaler  = StandardScaler()

X_train = x_scaler.fit_transform(X_train_df.values)
X_test  = x_scaler.transform(X_test_df.values)

yret_train = yret_scaler.fit_transform(Y_train_df[["y_ret"]].values).reshape(-1)
yret_test  = yret_scaler.transform(Y_test_df[["y_ret"]].values).reshape(-1)

yrv_train  = yrv_scaler.fit_transform(Y_train_df[["y_rv"]].values).reshape(-1)
yrv_test   = yrv_scaler.transform(Y_test_df[["y_rv"]].values).reshape(-1)
```

```
def make_sequences(X, y1, y2, lookback=30):
    Xs, y1s, y2s = [], [], []
    for i in range(lookback, len(X)):
        Xs.append(X[i-lookback:i, :])
        y1s.append(y1[i])
        y2s.append(y2[i])
    return np.array(Xs), np.array(y1s), np.array(y2s)

LOOKBACK = 30

Xtr, ytr_ret, ytr_rv = make_sequences(X_train, yret_train, yrv_train, lookback=LOOKBACK)
Xte, yte_ret, yte_rv = make_sequences(X_test,  yret_test,  yrv_test,  lookback=LOOKBACK)

print("Xtr:", Xtr.shape, "y_ret:", ytr_ret.shape, "y_rv:", ytr_rv.shape)
print("Xte:", Xte.shape, "y_ret:", yte_ret.shape, "y_rv:", yte_rv.shape)
```

```
def build_multitask_lstm(n_features, lookback):
    inp = layers.Input(shape=(lookback, n_features))

    x = layers.LSTM(64, return_sequences=True)(inp)
    x = layers.Dropout(0.2)(x)
    x = layers.LSTM(32)(x)
    x = layers.Dropout(0.2)(x)

    # Two heads
    out_ret = layers.Dense(1, name="ret")(x)
    out_rv  = layers.Dense(1, name="rv")(x)

    model = models.Model(inputs=inp, outputs=[out_ret, out_rv])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss={"ret": "mse", "rv": "mse"},
        loss_weights={"ret": 1.0, "rv": 1.0}
    )
    return model

model = build_multitask_lstm(n_features=Xtr.shape[-1], lookback=LOOKBACK)
model.summary()
```

```
es = callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

hist = model.fit(
    Xtr,
    {"ret": ytr_ret, "rv": ytr_rv},
    validation_split=0.2,
    epochs=200,
    batch_size=32,
    callbacks=[es],
    verbose=1
)
```
